```python
import pandas as pd
data=pd.DataFrame({'A':[1,2,4,1,2,4],
                  'B':[4,5,6,7,8,9],
                  'C':[0,0,0,0,0,0],
                  'D':[1,1,1,1,1,1]})
data.head()
```

```python
from sklearn.feature_selection import VarianceThreshold
var_thres=VarianceThreshold(threshold=0)
var_thres.fit(data)
```

```python
const_column=[column for column in data.columns
             if column not in data.columns[var_thres.get_support()]]
```

```python
print(len(const_column))
```

```python
for feature in const_column:
    print(feature)
```

```python
df=pd.read_csv(r"D:\Road to Data Science\DataSets Given In Class\Satendra.csv.csv",nrows=1000)
df.head()
```

```python
df.shape
```

```python
X=df.drop("TARGET",axis=1)
y=df["TARGET"]
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=0)
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.feature_selection import VarianceThreshold
var_thres=VarianceThreshold(threshold=0)
var_thres.fit(X_train)
```

```python
const_columns=[column for column in X_train.columns
             if column not in X_train.columns[var_thres.get_support()]]
print(len(const_columns))
```

```python
for feature in const_columns:
    print(feature)
```
